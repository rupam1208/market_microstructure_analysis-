```python
import numpy as np
from scipy import stats
import random
import matplotlib.pyplot as plt
```

# Brownian Motion

```python
T=1;nsteps=1000
nsample=10000
```

```python
dt=T/nsteps
```

```python
def brownian_path():
    W=[0]
    t=0;wk=0;
    while t<=T:
        wk+=np.sqrt(dt)*np.random.randn()
        W.append(wk)
        t+=dt
    return np.array(W)
```

```python
time_steps=np.linspace(0,T,nsteps+1)
```

```python
all_paths=np.array([brownian_path() for i in range(nsample)])
```

```python
mean_path=np.array([np.mean(all_paths[:,i]) for i in range(nsteps+1)])
```

```python
# consistency check
np.mean(all_paths[:,1000])
np.std(all_paths[:,1000])
```

```python
plt.plot(np.linspace(0,T,nsteps+1),all_paths[0])
plt.plot(np.linspace(0,T,nsteps+1),all_paths[1])
plt.plot(np.linspace(0,T,nsteps+1),all_paths[2])
plt.plot(np.linspace(0,T,nsteps+1),mean_path)
plt.xlabel("t")
plt.ylabel("W(t)")
```

```python
#probability of corssing a certain value
prob=(1/nsample)*np.array([len([i for i in all_paths[:,j] if i>1.5]) for j in range(1,nsteps+1)])
prob;
```
