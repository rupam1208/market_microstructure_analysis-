```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
x = np.linspace(0, 2 * np.pi, 200)
line, = ax.plot([], [])  # empty initial plot

def update(frame):
    # plot up to current frame
    line.set_data(x[:frame], np.sin(x[:frame]))
    return line,

ani = FuncAnimation(fig, update, frames=len(x), interval=30, blit=True, repeat=False)

ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.1, 1.1)

plt.show()
```
